```python
import numpy as np
import matplotlib.pyplot as plt

# Define layers (from our MNIST ANN)
layers = [784, 64, 10]  # Input, Hidden, Output
layer_names = ['Input', 'Hidden', 'Output']

# Plot structure
fig, ax = plt.subplots(figsize=(8, 6))
for i, (size, name) in enumerate(zip(layers, layer_names)):
    # Draw neurons as circles
    for j in range(min(size, 5)):  # Show up to 5 neurons per layer
        y = j - (min(size, 5) - 1) / 2  # Center vertically
        ax.add_patch(plt.Circle((i * 2, y), 0.3, fill=True, color='skyblue'))
    if size > 5:
        ax.text(i * 2, 0.5, '...', ha='center', va='center')
    ax.text(i * 2, -3, f'{name}\n({size})', ha='center', fontsize=12)

    # Draw connections (weights)
    if i < len(layers) - 1:
        for j in range(min(layers[i], 5)):
            y1 = j - (min(layers[i], 5) - 1) / 2
            for k in range(min(layers[i + 1], 5)):
                y2 = k - (min(layers[i + 1], 5) - 1) / 2
                ax.plot([i * 2 + 0.3, (i + 1) * 2 - 0.3], [y1, y2], 'k-', alpha=0.1)

# Style
ax.set_xlim(-1, len(layers) * 2 - 1)
ax.set_ylim(-4, 3)
ax.set_aspect('equal')
ax.axis('off')
plt.title("Deep Learning Structure (MNIST ANN)", fontsize=14)
plt.show()
```
